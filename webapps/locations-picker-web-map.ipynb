```python
# Application initialization
import mercury as mr

import warnings

warnings.filterwarnings("ignore")

show_code = mr.Checkbox(value=False, label="Show code", url_key="show_code")

app = mr.App(
    title="Locations Picker Web Map",
    description="A web map application for picking locations of interest and exporting to JSON file.",
    show_code=show_code.value,
    allow_download=False,
)
```

```python
import folium
from folium.plugins import Draw

m = folium.Map(
    location=(-0.624251, 38.414451),
    control_scale=True,
)

Draw(
    export=True,
    draw_options={
        "polyline": False,
        "polygon": False,
        "rectangle": False,
        "circle": False,
        "circlemarker": False,
    },
).add_to(m)

m
```
