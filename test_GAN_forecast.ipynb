```python
# Demo script to load and plot cGAN forecast data.

from datetime import datetime

# These are the three functions that you can use:
from show_forecasts.show_cGAN import load_GAN_forecast, plot_GAN_forecast, plot_GAN_ensemble
```

```python
# Where the forecasts are downloaded to
data_dir = '/home/c/cooperf/IFS/ICPAC_data/operational/GAN_forecasts'

# The date that the forecast was initialised
forecast_init_date = datetime(year=2024, month=4, day=19)
# Pick today instead:
#forecast_init_date = datetime.now()
```

```python
# Load the GAN forecast
data = load_GAN_forecast(forecast_init_date, data_dir)
```

```python
# Plot the GAN forecast
plot_GAN_forecast(data)
```

```python
# style options: None (default), 'ICPAC', 'ICPAC_heavy', 'KMD', 'EMI', 'EMI_heavy'
# plot_units can be 'mm/h' (default), 'mm/6h', 'mm/day' or 'mm/week'
# region can be 'ICPAC', 'Kenya', 'South Sudan', 'Rwanda', 'Burundi', 'Djibouti',
#               'Eritrea', 'Ethiopia', 'Sudan', 'Somalia', 'Tanzania', 'Uganda'
plot_GAN_forecast(data, style='ICPAC_heavy', plot_units='mm/day', region='Kenya')
```

```python
# Plot the full ensemble
plot_GAN_ensemble(data, valid_time_start_hour=12, style='ICPAC', plot_units='mm/day', region='Ethiopia')
```
