```python
# Demo script to load and plot cGAN forecast data.

from datetime import datetime

# These are the functions that you can use:
from show_forecasts.show_cGAN import load_GAN_forecast, plot_GAN_forecast, plot_GAN_ensemble
from show_forecasts.show_cGAN import plot_GAN_threshold_chance, plot_GAN_local_histograms, plot_location_marker
from show_forecasts.show_cGAN import sort_along_ensemble_axis, plot_GAN_exceedance
from show_forecasts.data_utils import print_locations
```

```python
# Where the forecasts are downloaded to
#data_dir = '/home/c/cooperf/IFS/ICPAC_data/operational/GAN_forecasts'
data_dir = '/Users/cooperf/Documents/WFP/data/ICPAC_operational/Jurre_brishti/forecasts'

# The date that the forecast was initialised
forecast_init_date = datetime(year=2024, month=5, day=21)
# Pick today instead:
#forecast_init_date = datetime.now()
```

```python
# Load the GAN forecast
data = load_GAN_forecast(forecast_init_date, data_dir)
```

```python
# Plot the GAN forecast
plot_GAN_forecast(data)
```

```python
# Plot just one time: Options are 6, 12, 18, 0 or 'all'.
# style options: None (default), 'ICPAC', 'ICPAC_heavy', 'KMD', 'EMI', 'EMI_heavy'
# plot_units can be 'mm/h' (default), 'mm/6h', 'mm/day' or 'mm/week'
# region can be 'ICPAC', 'Kenya', 'South Sudan', 'Rwanda', 'Burundi', 'Djibouti',
#               'Eritrea', 'Ethiopia', 'Sudan', 'Somalia', 'Tanzania', 'Uganda'
# A file name can be specified using file_name="my_file.pdf". File types can be
#               png, jpg or pdf.
#plot_GAN_forecast(data, valid_time_start_hour=12, style='KMD', plot_units='mm/6h', region='Kenya')
plot_GAN_forecast(data, valid_time_start_hour='all', style='KMD', plot_units='mm/6h', region='ICPAC')
```

```python
# Plot the rainfall over 24 hours
#plot_GAN_forecast(data, accumulation_time='24h', style='KMD',
#                  plot_units='mm/day', region='Kenya')
plot_GAN_forecast(data, accumulation_time='24h', style='KMD',
                  plot_units='mm/day', region='ICPAC')
```

```python
# Plot the full ensemble
#   valid_time_start_hour=6 - The hour the valid time starts at. Can either be 6, 12, 18 or 0 UTC.
#   max_num_plots=50        - Maximum number of ensemble members to plot.
plot_GAN_ensemble(data, valid_time_start_hour=18, style='KMD', plot_units='mm/6h', region='Kenya', max_num_plots=15)
```

```python
# Plot the chance of rainfall at a rate above a specified threshold.
#   threshold=2             - We'll plot the chance of rainfall above this threshold rate. The
#                             default is 2 mm/h. The units of threshold is set by plot_units.
#   plot_units='mm/h'       - Can be 'mm/h' (default), 'mm/6h', 'mm/day' or 'mm/week'
#   valid_time_start_hour=6 - The hour the valid time starts at. Can either be 6, 12, 18 or 0 UTC or 'all'.
#   style=None              - Can be 'ICPAC', 'KMD', 'EMI' or the default.
plot_GAN_threshold_chance(data, threshold=5, plot_units='mm/6h', valid_time_start_hour='all',
                          style='KMD', show_percentages=False, region='Kenya')
```

```python
plot_GAN_threshold_chance(data, threshold=2, plot_units='mm/6h', valid_time_start_hour='0',
                          show_percentages=False, region='Kenya')
```

```python
# Prints the named locations available for making histograms
# All locations are listed in the file show_forecasts/shapes/locations.py
#   Optionally restrict the country to one of 'Kenya', 'South Sudan', 'Rwanda',
#   'Burundi', 'Djibouti', 'Eritrea', 'Ethiopia', 'Sudan', 'Somalia',
#   'Tanzania', 'Uganda'
print_locations('Kenya')
```

```python
# Shows where the location is on the map
plot_location_marker(location_name="Andiwo's Village", region='Kenya', file_name=None)
```

```python
# Use a latitude and longitude to specify a custom location
plot_location_marker(location_name="Made up name", latitude=-1, longitude=38, region='Kenya', file_name=None)
```

```python
# The location name can be selected from the list returned by print_locations()
# All locations are listed in the file show_forecasts/shapes/locations.py
#   For the case of the same name in different countries, the country can optionally
#   be selected from 'Kenya', 'South Sudan', 'Rwanda', 'Burundi', 'Djibouti',
#   'Eritrea', 'Ethiopia', 'Sudan', 'Somalia', 'Tanzania', 'Uganda'.
# Be carefull. Sometimes there are a few places with the same name.
plot_GAN_local_histograms(data, 'Kirkos', country='Ethiopia', plot_units='mm/6h')
```

```python
# Or if you want to specify the latitude and longitude, you can call the location what you like
# Specify the number of equally spaced bins (columns) in the histogram.
# Plot a line indicating the amount of rain that will be exceeded with a given probability.
plot_GAN_local_histograms(data, 'Indian Ocean', latitude=-2.75, longitude=45.35, plot_units='mm/6h', num_bins=10,
                          probability=1/20)
```

```python
# Before plot_GAN_exceedance can be used, we must sort the data
# This might be slow, so we only want to do it once.
data_sorted = sort_along_ensemble_axis(data)
```

```python
# Plot the amount of rain that will be exceeded with a given probability.
plot_GAN_exceedance(data_sorted, probability=1/20, valid_time_start_hour='all', style='EMI',
                    plot_units='mm/6h', region='ICPAC', file_name=None)
```
